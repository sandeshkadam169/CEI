## 🩹 Step 3: Handle Missing Values

Missing values can cause errors in analysis. We need to:
1. **Identify** where missing values exist
2. **Decide** whether to fill or drop them

### 3.1 Identify Missing Values

```python
# Count missing values per column
missing_counts = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df) * 100).round(2)

# Create a summary table of missing values
missing_df = pd.DataFrame({
    'Missing Count': missing_counts,
    'Missing %': missing_percent
})

# Display only columns that have missing values
print("📋 Missing Values Summary:")
print("=" * 50)
missing_with_values = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)
print(missing_with_values)
print(f"\nTotal missing values in the dataset: {df.isnull().sum().sum()}")
```

```python
# Visualize missing values with a simple bar chart
if not missing_with_values.empty:
    print("\n📊 Missing Values Bar Chart:")
    print("-" * 50)
    for col, row in missing_with_values.iterrows():
        bar_length = int(row['Missing %'] / 2)
        bar = '█' * bar_length + '░' * (50 - bar_length)
        print(f"  {col:<25s} |{bar}| {row['Missing %']:5.1f}% ({int(row['Missing Count'])} nulls)")
```

### 3.2 Handle Missing Values — Fill & Drop Strategy

**Strategy:**
- **`discount`** (numerical, 121 missing) → Fill with **0** (no discount applied)
- **`seller_name`** (text, 301 missing) → Fill with **"Unknown"**
- **`what_customers_said`** (text, 573 missing) → Fill with **"No reviews"**
- **`videos`** (text, 781 missing) → Drop this column entirely (>78% missing, not useful for analysis)
- **`seller_information`** (text, 301 missing) → Fill with **"Not Available"**
- **`variations`** (text, 562 missing) → Fill with **"No variations"**

```python
# --- FILL missing values ---

# Fill 'discount' with 0 (no discount)
df['discount'] = df['discount'].fillna(0)
print(f"'discount' — filled NaN with 0")

# Fill 'seller_name' with 'Unknown'
df['seller_name'] = df['seller_name'].fillna('Unknown')
print(f"'seller_name' — filled NaN with 'Unknown'")

# Fill 'what_customers_said' with 'No reviews'
df['what_customers_said'] = df['what_customers_said'].fillna('No reviews')
print(f"'what_customers_said' — filled NaN with 'No reviews'")

# Fill 'seller_information' with 'Not Available'
df['seller_information'] = df['seller_information'].fillna('Not Available')
print(f"'seller_information' — filled NaN with 'Not Available'")

# Fill 'variations' with 'No variations'
df['variations'] = df['variations'].fillna('No variations')
print(f"'variations' — filled NaN with 'No variations'")
```

```python
# --- DROP columns with too many missing values ---

# Drop 'videos' column (>78% missing, not useful for analysis)
df = df.drop(columns=['videos'])
print(f"✅ 'videos' column dropped (>78% values were missing)")
```

```python
# Verify: No more missing values
remaining_missing = df.isnull().sum().sum()
print(f"\nRemaining missing values: {remaining_missing}")
if remaining_missing == 0:
    print("All missing values have been handled successfully!")
else:
    print("Some missing values still remain:")
    print(df.isnull().sum()[df.isnull().sum() > 0])
```

Demonstrate key Pandas operations:

## 🧮 Step 6: Create a Derived Column

**Task:** Create `total_amount = price × quantity`

Since our dataset doesn't have a direct `quantity` column (each row is a single product listing), we adapt this step as follows:

1. **`discounted_price`** = `initial_price × (1 - discount/100)` — the actual price after discount
2. **Simulated `quantity`** — We simulate a `quantity` column (random values 1-5) to demonstrate the formula
3. **`total_amount`** = `discounted_price × quantity`

```python
# Step 6a: Create 'discounted_price' column from initial_price and discount
df['discounted_price'] = (df['initial_price'] * (1 - df['discount'] / 100)).round(2)

print(" Created 'discounted_price' column")
print(f"   Formula: discounted_price = initial_price × (1 - discount/100)")
print()
print(df[['product_id', 'title', 'initial_price', 'discount', 'discounted_price']].head(10))
```

```python
# Step 6b: Simulate a 'quantity' column (1 to 5 units)
np.random.seed(42)  # For reproducibility
df['quantity'] = np.random.randint(1, 6, size=len(df))

print(" Created 'quantity' column (simulated: random values 1–5)")
print(f"   Quantity distribution:")
print(df['quantity'].value_counts().sort_index())
```

```python
# Step 6c: Create 'total_amount' = discounted_price × quantity
df['total_amount'] = (df['discounted_price'] * df['quantity']).round(2)

print("✅ Created 'total_amount' column")
print(f"   Formula: total_amount = discounted_price × quantity")
print()
print(df[['product_id', 'title', 'initial_price', 'discount', 
          'discounted_price', 'quantity', 'total_amount']].head(10))
```

```python
# Summary statistics of the new derived columns
print("Summary Statistics of Derived Columns:")
print("=" * 50)
print(df[['initial_price', 'discount', 'discounted_price', 'quantity', 'total_amount']].describe().round(2))
```
